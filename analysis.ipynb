```python
import json
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# read the json file /ceph/tsesterh/abstraction/trove/output_codellama/CodeLlama-7b-Instruct-hf/math/algebra/results/Noneex/trove/run3.json
with open('/ceph/tsesterh/abstraction/trove/output_codellama/CodeLlama-7b-Instruct-hf/math/algebra/results/Noneex/trove/run2.json') as f:
    data = json.load(f)
```

```python
# print(data[3]['mode'])
# print(data[3]['response']['response'])

for entry in data:
    if entry['mode'] == "create" :
        print("---------\nNew Entry: ")
        print(entry['response']['response'])
```

```python
is_correct_counts = Counter(
    entry['mode'] for entry in data if entry['response']['is_correct']
)

# Extract data for plotting
modes = list(is_correct_counts.keys())
counts = list(is_correct_counts.values())

# Plotting
plt.figure(figsize=(8, 6))
plt.bar(modes, counts)
plt.xlabel('Mode')
plt.ylabel('Count of is_correct=True')
plt.title('Count of is_correct=True per Mode')
plt.show()
```
